```python
# URL of page to be scraped
#Scrape the NASA Mars News Site and collect the latest News Title 
#and Paragraph Text. Assign the text to variables that you can reference later.

#first set the URL
url = 'https://mars.nasa.gov/news/'

#retrieve the page
response = requests.get(url)

#check the response code
response
```

```python
#create a BeautifulSoup object and parse the page
soup = BeautifulSoup(response.text, 'html.parser')

#pretty print the results
print(soup.prettify())
```

```python
#get just the title
news_title = soup.find('div', class_='content_title').text
#we are going to put our results in a variable called news_title, and we are going to fetch the text of the 'content_title' div tag
news_title
```

```python
#then we are going to drop the Newline characters
clean_title = str(news_title)
print (clean_title)
```

```python
#Fetch a paragraph
#it's of the class 'rollover_description_inner'
news_p = soup.find('div', class_='rollover_description_inner').text
news_p
```

```python
#drop the Newline's and other formatting
clean_p = str(news_p)
print (clean_p)
```

```python
#Scrape the JPL webpage for the current Mars image
#this requires the Chromedriver executable to be on the search path or in the same directory as the script
browser = Browser('chrome')
url_img = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url_img)
html = browser.html
soup_div = BeautifulSoup(html,'html.parser')
print(soup_div.prettify())
```

```python
#From the larger results, select the img object, thumb class (since that is how it's defined in the web page code)
image_results = soup_div.find('img', class_="thumb")
print(image_results)
```

```python
#Extract the relative file path from this string
image_src = image_results['src']
print(image_src)
```

```python
#Now append the rest of the URL string
jpl_featured_image = 'https://www.jpl.nasa.gov/' + image_src
print(jpl_featured_image)
```

```python
#Mars Hemispheres
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html,'html.parser')
hemispheres = soup.find('div',class_='collapsible results')
results = hemispheres.find_all('a')
print(results)
```

```python
#count how many results we have
len(results)
```

```python
#create an empty list
hemispheres = []
#populate it with the images data
for result in results:
    if result.h3:
        title = result.h3.text
        link = 'https://astrogeology.usgs.gov' + result['href']
        print(title,link)    
        browser.visit(link)
        time.sleep(5)
        image_html = browser.html
        soup = BeautifulSoup(image_html,'html.parser')
        soup_image = soup.find('div', class_='downloads').find('li').a['href']
        print(soup_image)
        mars_images = {'title':title, 'img_url':soup_image}
        hemispheres.append(mars_images)
```

```python
#print the list

print(hemispheres)
```
